### Sentiment Analysis with Vader 
Notes (11-29-17):
Strategies:
- look at the asunto as a whole and pass that to the analyser 
- look at each sentence within the asunto, and pass each of those to the analyser, and take the average across the asunto

```python
analyser = SentimentIntensityAnalyzer()
```

```python
def print_sentiment_scores(sentence):
    snt = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(snt)))
```

```python
sentences = [
                "The plot was good, but the characters are uncompelling and the dialog is not great.", 
                "A really bad, horrible book.",       
                "At least it isn't a horrible book."
            ]
for sentence in sentences:
    print_sentiment_scores(sentence)
```
